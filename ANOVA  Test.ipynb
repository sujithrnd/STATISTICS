# One-Way ANOVA in Python: 
Hypothesis involved:
one-way ANOVA has the below given null and alternative hypotheses:

H0 (null hypothesis): μ1 = μ2 = μ3 = … = μk (It implies that the means of all the population are equal)
H1 (null hypothesis): It states that there will be at least one population mean that differs from the rest

```python


#Scenario:Researchers took 20 cars of the same to take part in a study. These cars are randomly doped with one of 
#the four-engine oils and allowed to run freely for 100 kilometers each. At the end of the journey, the performance of 
#each of the cars is noted
# H0 -> No difefrence in performance for teh cars
# H1 ->There are difference in performance depending on the oil
```

```python
# Performance when each of the engine
# oil is applied
performance1 = [89, 89, 88, 78, 79]
performance2 = [93, 92, 94, 89, 88]
performance3 = [89, 88, 89, 93, 90]
performance4 = [81, 78, 81, 92, 82]
```

```python
# Importing library
from scipy.stats import f_oneway
# Conduct the one-way ANOVA
f_oneway(performance1, performance2, performance3, performance4)
```

```python
 #Since the p-value is less than 0.05 hence we would reject the null hypothesis. 
#This means we have sufficient proof to say that there exists a difference in the performance among four different engine oils. 
```

# Two-Way ANOVA : 
scientists need to know whether plant growth is affected by fertilizers and watering frequency. They planted exactly 30 plants and allowed them to grow for six months under different conditions for fertilizers and watering frequency. After exactly six months, they recorded the heights of each plant centimeters. 

H0 -> Null hypthesys ->Watering fertilizer & watering frequency has no effect

H1-> Alternate hypothesys - > There is an effect of watring frequency & fertilizer has an effect on plant growth

```python
# Importing libraries
import numpy as np
import pandas as pd
# Create a dataframe
dataframe = pd.DataFrame({'Fertilizer': np.repeat(['daily', 'weekly'], 15),
'Watering': np.repeat(['daily', 'weekly'], 15),
'height': [14, 16, 15, 15, 16, 13, 12, 11, 14,15, 16, 16, 17, 18, 14, 13, 14, 14,14, 15, 16, 16, 17, 18, 14, 13, 14,14, 14, 15]})
```

```python
dataframe
```

```python
# Importing libraries
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Create a dataframe
dataframe = pd.DataFrame({'Fertilizer': np.repeat(['daily', 'weekly'], 15),
                          'Watering': np.repeat(['daily', 'weekly'], 15),
                          'height': [14, 16, 15, 15, 16, 13, 12, 11, 14, 
                                     15, 16, 16, 17, 18, 14, 13, 14, 14, 
                                     14, 15, 16, 16, 17, 18, 14, 13, 14, 
                                     14, 14, 15]})


# Performing two-way ANOVA
model = ols('height ~ C(Fertilizer) + C(Watering) +\
C(Fertilizer):C(Watering)',
data=dataframe).fit()
result = sm.stats.anova_lm(model, type=2)

# Print the result
print(result)
```

# Result
C(Fertilizer):C(Watering)  p value 0.886 >significance 0.05 so accepting null hypothesys means these 2 factors has no effect on plant growth
